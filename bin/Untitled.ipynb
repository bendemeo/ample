```python
def grow(data, subsets = None, color=None):
    
    
    if subsets is None:
        subsets = data.subsets.keys()
    
    if color is None:
        color = data.annos.keys()
    
    max_n = max([e.shape[0] for e in [data.embeddings[name] for name in embedding_types]])
    print(max_n)
    
    @interact
    def build_plot(N=(1,max_n,1), method = embedding_types, color=color):
        if color is None:
            colors = [1]*data.numObs
        else:
            le = LabelEncoder().fit(data.annos[color])
            colors = le.transform(data.annos[color])
        
        numPts = min([N, data.embeddings[method].shape[0]])
        
        plt.scatter(data.embeddings[method][:N,0], 
                    data.embeddings[method][:N,1],
                    label = [data.annos[color][i] for i in data.subsamples[method][:N]],
                   c=[colors[i] for i in data.subsamples[method][:N]],
                   cmap='Set1')
        plt.legend()
            
    
```
